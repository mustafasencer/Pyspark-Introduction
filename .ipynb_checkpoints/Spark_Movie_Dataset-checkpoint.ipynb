```python
rdd2 = sc.textFile("C:\\Users\\lenovo\\Desktop\\PROJECTS\\SPARK\\popculture-imdb-5000-movie-dataset\\data\\movie_metadata.csv")
```

```python
header = rdd2.first()
rdd2 = rdd2.filter(lambda x: x != header)
rdd2 = rdd2.map(lambda x: x.split("http"))
rdd2 = rdd2.map(lambda x: (x[1].split(",")[3], float(x[1].split(",")[8])))
```

```python
country_point = rdd2.combineByKey((lambda x: (x, 1)),
                                 (lambda x, y: (x[0] + y, x[1] +1)),
                                 (lambda x, y: (x[0] + y[0], x[1] + y[1])))

country_avg = country_point.map(lambda country: (country[0], country[1][0]/country[1][1]))
```

```python
country_avg.sortBy(lambda x: -x[1]).collect()
```

What is the analysis concerning the directors

```python
rdd3 = sc.textFile("C:\\Users\\lenovo\\Desktop\\PROJECTS\\SPARK\\popculture-imdb-5000-movie-dataset\\data\\movie_metadata.csv")
```

```python
header = rdd3.first()
rdd3 = rdd3.filter(lambda x: x != header)
rdd3 = rdd3.map(lambda x: x.split("http"))
rdd3 = rdd3.map(lambda x: (x[0].split(",")[1], float(x[1].split(",")[-3])))
```

```python
director_point = rdd3.combineByKey(lambda x: (x, 1),
                                 (lambda x, y: (x[0] + y, x[1] +1)),
                                 (lambda x, y: (x[0] + y[0], x[1] + y[1])))
director_ortalama_puan = director_point.map(lambda x: (x[0], x[1][0] /x[1][1]))
director_ortalama_puan.sortBy(lambda x: -x[1]).collect()
```

```python
director_toplam_puan = director_point.map(lambda x: (x[0], x[1][0]))
director_toplam_puan.sortBy(lambda x: -x[1]).collect()                                    
```

```python
director_film_sayısı = director_point.map(lambda x: (x[0], x[1][1]))
director_film_sayısı.sortBy(lambda x: -x[1]).collect()
```

Does big budget mean big box-office

```python
rdd5 = sc.textFile("C:\\Users\\lenovo\\Desktop\\PROJECTS\\SPARK\\popculture-imdb-5000-movie-dataset\\data\\movie_metadata.csv")
```

```python
rdd5 = rdd5.filter(lambda x: x!= header)
rdd5 = rdd5.map(lambda x: x.split("http")).filter(lambda x: x[0].split(",")[8] != "").filter(lambda x: x[1].split(",")[5] != "")
budget = rdd5.map(lambda x: (x[0].split(",")[11], int(x[1].split(",")[5])))
```

```python
budget.sortBy(lambda x: -x[1]).collect()
```

```python
gross = rdd5.map(lambda x: (x[0].split(",")[11], int(x[0].split(",")[8])))
```

```python
gross = gross.sortBy(lambda x: -x[1])
```

```python
say = 0

for i in gross.take(20):
    say += 1
    print(say, i)
```
